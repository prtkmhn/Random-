```python
from __future__ import annotations
import os
import uuid
import gradio as gr
import torch
from diffusers import PixArtAlphaPipeline

# Check for CUDA availability for GPU acceleration
if not torch.cuda.is_available():
    raise EnvironmentError("This demo requires a GPU to run.")

# Set device to GPU
device = torch.device("cuda:0")

# Load the pipeline
pipe = PixArtAlphaPipeline.from_pretrained(
    "PixArt-alpha/PixArt-XL-2-512x512",
    torch_dtype=torch.float16,
    use_safetensors=True,
).to(device)

# Function to save generated image
def save_image(img):
    unique_name = str(uuid.uuid4()) + ".png"
    img.save(unique_name)
    return unique_name

# Main function to generate images
def generate(prompt: str):
    if prompt.strip() == "":
        return None  # No prompt provided, return nothing
    images = pipe(
        prompt=prompt,
        width=512,  # Adjust width as needed
        height=512,  # Adjust height as needed
        num_inference_steps=4,  # Adjust steps as needed for faster generation or better quality
        output_type="pil",
    ).images

    image_paths = [save_image(img) for img in images]
    return image_paths[0]  # Return the first image path

# Gradio interface
with gr.Blocks() as demo:
    with gr.Row():
        prompt = gr.Textbox(
            label="Type here:",
            placeholder="Start typing...",
            lines=2,
            max_lines=5,
            live=True,  # Update as user types
        )
    result = gr.Image(label="Generated Image")

    prompt.change(fn=generate, inputs=prompt, outputs=result)

if __name__ == "__main__":
    demo.launch(share=True)
```
